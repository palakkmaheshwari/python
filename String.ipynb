```python
test1 = [0,'pavi','palak',9999,'vivek','Yogi',9998,7,65,'A']
```

```python
for i in test1:
    if test1.index(i)%2 != 0:
        if str(i).isnumeric() == False and len(str(i)) == 4:
            print(i,'---> Soni')
        elif str(i).isnumeric() == False:
            print(i, 999)
        elif str(i).isnumeric() == True and i%3 == 0:
            print(i, 'Lucky')
        else:
            print(i)
    else:
        print(i)
```

```python
test22 = [180,'palak',360,8,'alak','pavi', 'p']
```

```python
test22_op = []
test22_nam = []

for i in test22:
    if str(i).isnumeric() == False:
        if len(str(i)) == 4:
            s = i.capitalize()
            test22_nam.append(s)
        elif len(str(i)) != 4:
            test22_nam.append(i[::-1])
    else:
        test22_op.append(i)
        
test22_op.sort()
test22_nam.sort(key=len)
test22_op.extend(test22_nam)
```

```python
god = ['Rama','Sita','Ganesh','Krishna','Shiv','Sati','Kartikeya','palak']
```

```python
gods = []

for i in god:
    if len(i) == 4 and i[-1] != 'a':
        s = i.lower()
        gods.append(s[::-1])
    elif i.startswith('K') and i.endswith('a'):
        gods.append(len(i))
    elif i.endswith('a'):
        f = i.upper()
        gods.append(f)
    elif i.startswith('Ga') and i.endswith('sh'):
        gods.append(i.swapcase())
    else:
        gods.append('Oops')     
    
```

```python
'''
Ques-

Condition 1: print 'buzz' if number is divisible by 3.
Condition 2: print 'duzz' if number is divisible by 7.
Condition 3: print 'buzzduzz' if number is divisible by both 3 and 7.

'''

def buzzduzz():
    n = input('Please enter an interger - \n')
    if n.isnumeric() == True:
        n = int(n)
        if n%3 == 0 and n%7 == 0:
            print('buzzduzz')
        elif n%7 == 0:
            print('duzz')
        elif n%3 == 0:
            print('buzz')
        else:
            print('Entered input is not divisible by 3,7 or both.')
    else:
        print('Please enter a valid input!')
```

```python
buzzduzz()
```

```python
'''
Ques-

Condition 1: Take a string input from the user.
Condition 2: If input has more than 7 vowel characters, in that case ignore all the last characters of the words in the string 
and return the output.
Condition 3: If the start and end of the string are not vowels then reverse the string.
Condition 4: If the end of the string is vowel then return last character only.
Condition 5: If string has more than 10 words then print 'Too Long'.
Condition 6: If string contains exact length 10 then print 5th and 4th word.
Condition 7: Else print the input as it is.

'''

def stringchecker():
    n = input('Please enter a string - \n')
    if len(n.split()) == 10:
        print(' '.join(list(reversed([i for i in n.split()][3:5]))))
    elif len(n.split()) > 10:
        print('Too Long')
    elif n[0].lower() not in 'aeiou' and n[-1].lower() not in 'aeiou':
        return ' '.join(list(reversed(n.split())))
    elif len([i for i in n if i in 'aeiou']) > 7:
        return [i[:-1] for i in n.split()]
    elif n[-1].lower() in 'aeiou':
        print(n[-1])
    else:
        print(n)    
```

```python
stringchecker()
```
