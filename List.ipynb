```python
l = [[[[55]],7],1,2,3,[[[22,23]]],[4,5,6],[[22,33],7,[80,90,[1,2,3]]]]

def flatten(l):
    while list in [type(i) for i in l]:
        a = []
        for i in l:
            if type(i) == int:
                a.extend([i])
            else:
                a.extend(i)
        l = a
    return a

%timeit flatten(l)
```

```python
def flatten_list(lst):
    result = []
    for i in lst:
        if isinstance(i, list):
            result.extend(flatten_list(i))  # Recursively flatten nested lists
        else:
            result.append(i)
    return result

l = [[[[55]],7],1,2,3,[[[22,23]]],[4,5,6],[[22,33],7,[80,90,[1,2,3]]]]
l = flatten_list(l)
print(l)
%timeit flatten_list(l)
```

```python
def flatten_list(lst):
    result = []
    stack = lst.copy()  # Create a copy to avoid modifying the original list
    
    while stack:
        item = stack.pop(0)  # Pop the first element
        if isinstance(item, list):
            stack = item + stack  # Prepend items to be processed first
        else:
            result.append(item)
    
    return result

l = [[[[55]],7],1,2,3,[[[22,23]]],[4,5,6],[[22,33],7,[80,90,[1,2,3]]]]
l = flatten_list(l)
print(l)
%timeit flatten_list(l)
```
