# Dictionary Comprehension
`It is used with for-loop to iterate over the iteraters within {} brackets, it will take key-value pair. It gives the output in the form of dictionary.`<br>
`Without a key-value pair it will be called as a set comprehension.`

```python
# List comprehension

[i**2 for i in range(11)]
```

```python
# Check type

type([i**2 for i in range(11)])
```

```python
# Set comprehension

{i**2 for i in range(11)}
```

```python
type({i**2 for i in range(11)})
```

```python
# Dictionary comprehension

{i:i**2 for i in range(11)}
```

```python
type({i:i**2 for i in range(11)})
```

```python
# It is a generator and retrive the output of uniques values of iterable we use set

set(i**2 for i in [5,5,6,7,8,8,0])
```

```python
# It is a generator

(i**2 for i in range(11))
```

```python
type(i**2 for i in range(11))
```

```python
[i for i in (i**2 for i in range(11))]
```

```python
# Convert the output of LC in tuple

tuple([i**2 for i in range(11)])
```

```python
a = (i**2 for i in range(11))
a
```

```python
# For loop is use to retrive the output 

for i in a:
    print(i)
```
